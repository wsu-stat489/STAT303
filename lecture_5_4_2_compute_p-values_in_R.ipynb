# $p$-values for the Binomial Distribution

In this lecture, we will

* Estimate $p$-values with a simulation.
* Compute exact $p$-values with `dbinom` and `pbinom`
* Use $p$-values to decide if a value is unusual.

## Motivating examples.

Recall the Lebron James is a 73.4% FT shooter.  We will imagine that he is going to shoot 300 free throws.

This will have a binomial distribution because

1. Fixed $n$ and $p$
2. Assumed independence.

## Simulating the distribution

Let's simulate and plot the resulting distribution.

```r
n <- 300
p <- 0.734
trials <- rbinom(100000, n, p)
hist(trials, nclass = 500)
```

## Estimating one-tail $p$-values

We can estimate a one tail p-value using 

1. an inequality
2. the `mean` function

#### Lower $p$-value of 200

```r
mean(trials <= 200)
```

#### Upper $p$-value of 230

```r
mean(trials >= 230)
```

## Computing Exact $p$-values using `dbinom`

We can find the exact $p$-value is using `dbinom` and `sum` over the right range of values.

#### Lower $p$-value of 200

```r
sum(dbinom(0:200, n, p))
```

#### Upper $p$-value of 230

```r
sum(dbinom(230:n, n, p))
```

## Determining if a value is unusual

Recall that

* `"unusually large"` $\longrightarrow$ upper/right-tail
* `"unusually small"` $\longrightarrow$ upper/left-tail


#### Example 1 - Would 250 out of 300 made free-throws be an unusually large?

```r
# unusually large ==> upper/right-tail.
mean(trials >= 250)
```

> Yes, since the $p$-value $\approx$ 0 < 0.05, 250 out of 300 FT made would be unusually large

#### Example 2 - Would 210 out of 300 made free throws by unusually small?

```r
sum(dbinom(0:210, 300, 0.734))
```

> No, as the $p$-value = 0.104 > 0.05, this is not unusually small number of made FT.
